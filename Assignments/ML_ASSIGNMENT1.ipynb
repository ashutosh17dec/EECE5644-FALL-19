# Solution for 4.2:

```python

import numpy as np
import scipy.stats as stats
import math

mean=1
sigma=math.sqrt(2)
mean1=0
sigma1=1
y = np.linspace(-10,10)
x = np.linspace(-10,10)

plt.figure(1, figsize=(10, 6), dpi=100, facecolor='w', edgecolor='k')
x1 = stats.norm.pdf(x, mean, sigma)
x2 = stats.norm.pdf(x, mean1, sigma1)
plt.plot(x, x1, label='mean=1,variance=2', color = "red")
plt.plot(x, x2, label='mean=0,variance=1', color = "purple")
plt.plot(0.840, 0.28, marker = '^')
plt.plot(-2.8401, 0.007, marker = '^',color = "green")
plt.ylabel('P(x|L)')
plt.xlabel('x')
plt.title('Class conditioned pdf')
plt.legend()
plt.show()


plt.figure(2, figsize=(10, 6), dpi=100, facecolor='w', edgecolor='k')
x1 = stats.norm.pdf(x, mean, sigma)/((stats.norm.pdf(x, mean, sigma)+stats.norm.pdf(x, mean1, sigma1)))
x2 = stats.norm.pdf(x, mean1, sigma1)/((stats.norm.pdf(x, mean, sigma)+stats.norm.pdf(x, mean1, sigma1)))
plt.plot(y,x1, label='Posterior for L=2', color = "red")
plt.plot(y,x2, label='Posterior for L=1', color = "purple")
plt.legend()
plt.title('Posterior Probability')
plt.ylabel('P(L|x)')
plt.xlabel('x')
plt.show()
```
